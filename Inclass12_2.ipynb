```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import re
```

```python
r = requests.get("http://www.pogo.com")
data = r.text
```

```python
soup = bs(data)
```

```python
soup.title
```

```python
soup.body
```

```python
soup.a
```

```python
soup.find_all("a")
```

```python
soup.body.parent
```

```python
soup.find_all('a')
```

```python
soup.find_all(href="http://www.pogo.com/board-games")
```

```python
soup.find(text=re.compile('Monopoly'))
```

```python
soup.find_all(class_="card")
```

```python
soup.find_all(id=True)
```

```python
l1 = [link.get('href') for link in soup.find_all('a')]
l1
```

```python
pa = pd.DataFrame(l1, columns = ['urls'])
pa
```
